```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt 
import joblib
model = joblib.load('NB-digitclassification_after_Adding_images')
```

```python
#import PIL
from PIL import Image , ImageOps
img1 = Image.open("./8digit.png").convert('L')
im_invert = ImageOps.invert(img1)
im_invert.save('8_invert.png')
plt.imshow(im_invert, cmap= 'gray')
plt.show()
```

```python

new_img=im_invert.resize((8,8) , resample = Image.Resampling.BICUBIC)
#img.thumbnail((8,8) , resample = Image.Resampling.BICUBIC)
img=np.array(new_img)# thumbnail function is in pillow library
print(img.shape)
```

```python
plt.imshow(img,cmap='gray');
```

```python
# image_flatten = img.flatten()  # Flatten the array
image1 = img.reshape(1,-1)
print(image1.shape)
plt.imshow(image1 , cmap = 'gray');
#plt.show()
```

```python
predicted_digit = model.predict(image1)

print("predicted digit :", predicted_digit )
```
